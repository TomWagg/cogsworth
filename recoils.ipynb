```python
import cogsworth
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
q_range = np.linspace(1e-5, 1.0, 1000)
m_big = 10
m_small = q_range * m_big

plt.plot(q_range, cogsworth.kicks.get_recoil_kick(m_small, m_big), lw=3)
plt.xlabel("Mass ratio")
plt.ylabel("Recoil kick [km/s]")
plt.grid()
plt.show()
```

```python
q_range = np.linspace(1e-5, 1.0, 1000)
m_big = 10
m_small = q_range * m_big

plt.plot(q_range, cogsworth.kicks.get_recoil_kick(m_small, m_big, 0.1, 0.1), lw=3)
plt.xlabel("Mass ratio")
plt.ylabel("Recoil kick [km/s]")
plt.grid()
plt.show()
```
